```python
import numpy as np
from matplotlib.pylab import *

#x = float(input('Enter value of x less than 1: '))

x = 0.6

def Exact_Value(x):
    Val = 1/(1-x)
    return Val 

#N = int(input('Enter number of terms: '))

N = 70

Taylor = np.zeros(N)

for i in range(N):
    Taylor[i] = x**i
    Taylor[i] += Taylor[i-1]

    
#Create list of the exact value using the function

Exact_List = np.ones(N) * Exact_Value(x)

x2 = np.arange(0, N)

#Set up graph 1 

fig1 = figure(figsize=(8,8));
title('Taylor Series vs Order', fontsize=25)
grid() #turns grid on the graph on 
xlabel('Order Number', fontsize=25)
ylabel('Taylor Series Value', fontsize=25)
plt.yticks(arange(1,2.51,.25))
rcParams.update({'font.size': 20})

plt.plot(x2,Taylor, '-or',label = 'Taylor Series')
plt.plot(x2,Exact_List, label = 'Exact Value (2.5)', linewidth = 2)
legend(loc='best',fontsize=20)

'''As shown by the graph below, the minimum order of the Taylor Expansion needed to recover the exact value
of the function is roughly 12, this can be shown with the command Taylor[11]'''

print(Taylor[11])

errors = np.zeros(N)

for n in range(N):
    errors[n] = abs(Taylor[n] - Exact_List[0])/(Exact_List[0])

fig2 = figure(figsize=(8,8));
plt.yscale('log')
title('Taylor Series vs Order', fontsize=25)
grid() #turns grid on the graph on 
xlabel('Order Number', fontsize=25)
ylabel('Relative Error', fontsize=25)
rcParams.update({'font.size': 20})

plt.plot(x2, errors, '-o')
plot(69,errors[69],'or')

#The error becomes less than 10^-16 at around the 70th term as shown in the log plot 
```
