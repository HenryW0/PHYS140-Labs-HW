```python
# Number 2a) 

#Set up constants: 
# G = 6.67 × 10^−11 m^3 kg^−1 s^−2 (Gravitational Constant)
# M = 5.97 × 10^24 kg (Mass of Earth)
# R= 6371 km (Radius)

'''Since the orbit is circular, we can use uniform circular motion and the centripetal force equation F_c = m_s*v^2/r
This centripetal force is the gravitational force between Earth and the satellite therefore we set them equal
G * M_e * m_s/r^2 = m_s*v^2/r we can then cancel out like terms such as radius of Earth and the mass of the satellite
This simplfies to G * M_e /r = v^2, where the speed is equal to distance over time which in this case is the circumference and
orbital period, and since C = 2 * pi * r, v = 2 * pi * r / T, rearranging this equation we obtain: 
G * M_e = rv^2 which is r * ((2* pi * r)/T) ^ 2 thus  G * M_e = (4 * pi^2 * r^3) /T^2. Realizing that r is the height of the
satallite from Earth's center we can isolate and solve for r as this is the quantity that we want, so isolate r such that 
r^3 = (G * M_e * T^2) / (4 * pi^2), cube rooting both sides give us r = ((G * M_e * T^2) / (4 * pi^2)) ^ (1/3)
the last step is to substitute R_E + h for r since r already contains the radius of the Earth and height of the satellite
above the surface. Finally this gives us the equation that we want h = ((G * M_e * T^2) / (4 * pi^2)) ^ (1/3) - R_E'''

#Part b) 

#Define constants used in orbital height equation in correct units

G = 6.67 * 10**(-11)
M = 5.97 * 10**(24)
R = 6371 * 1000 

#First step is to ask user for input of the orbital period (T) 

T = eval(input('Enter orbital period (s): '))

#Next define the equation to calculate and print the height with all required constants

def height(T):
    h = (((G * M * (T)**(2))/(4 * (np.pi)**2))**(1/3)) - R 
    return h

print(f'Altitude of Satellite in meters: {height(T)}', 'm')

print('') #Space between prints

#Part c)

#Calculate geosynchronous orbit where T = 60 * 60 * 24 = 86400 s

print(f'Altitute for geosynchronous orbit in meters: {height(86400)}', 'm')

#Calculate 90 minute orbit T = 90 * 60 = 5400 s

print(f'Altitute for 90-minute orbit in meters: {height(5400)}', 'm')

#Calculate 45 minute orbit T = 45 * 60 = 2700 s

print(f'Altitute for 45-minute orbit in meters: {height(2700)}', 'm')

'''From these calculations we can conclude that both the geosynchronous orbit and 90-minute orbit require the satellite to
be above the Earth's surface due to their slow speed whereas the 45-minute orbit satellite is travelling much faster
where it technically could be underneath the surface and still be able to orbit the Earth if that was possible'''

print('')

#Part d)

TS = 86148 #sidereal day time in seconds 23.93 * 60 * 60 = 86148s
TD = 86400 #Normal 24 hour day

print(f'Altitude of Sidereal Satellite in meters: {height(TS)}', 'm')
print(f'Altitude of 24-Hour Satellite in meters: {height(TD)}', 'm')
print(f'Difference between altitudes in meters: {height(TD) - height(TS)}', 'm')

'''The reason for this is because the satellite's position is taken with respect to a distant star whereas in the extra
3 minutes 56 seconds the satellite would facing our sun again in the same position instead of the distant star due 
to the rotation and orbit. For this reason a sidereal day is shorter than a 24 hour day '''

print() #Printing here just because otherwise it will print out my above comment since it is the last input
```

```python
#Set up constants for quantum potential step

mass_electron = 9.11e-31 #kilograms
energy_electron = 10 #eV electron volts
potential_step = 9 #eV potential step height
reduced_planck = 6.58211956e-16 #in units of eV⋅s

#Calculate wavevectors first

k_1 = np.sqrt(2 * mass_electron * energy_electron)/reduced_planck
k_2 = np.sqrt(2 * mass_electron * (energy_electron - potential_step))/reduced_planck

#Calculate probabilities corresponding to transmission and reflection 

prob_T = (4 * k_1 * k_2)/((k_1 + k_2)**2) #probability for transmission
prob_R = ((k_1 - k_2)/(k_1 + k_2))**2 #probability for reflection

print(f'The probability that the electron passes the step is: {prob_T * 100}','%') #multiply values by 100 and add % 
print(f'The probability that the electron gets reflected is: {prob_R * 100}','%')
```

```python
#Planetary Orbits Part a)

'''Given the equation (v_2)^2 - (2GM /(v_1 * l_1)) v_2 - [(v_1)^2 - 2GM/l_1] = 0 since this is a quadratic equation we can solve
for v_2 where the solution is v_2 = ((2GM /(v_1 * l_1)) +/- sqrt((2GM /(v_1 * l_1))^2 + 4[(v_1)^2 - 2GM/l_1]))/2
The inside of the squareroot becomes 4G^2 M^2/v_1^2 * l_1^2 + 4 v_1^2 -8GMl_1 then we can factor out sqrt(4/v_1^2) which
leaves us with G^2 M^2/l_1^2 - 2GM/l_1 *v_1^2 +v_1^4 under the squareroot. Recognizing that this is the form a^2 -2ab +b^2
we can factor inside and get GM/l_1 - v_1^2, simplfying we then reach ((GM /(v_1 * l_1)) +/- GM/l_1* v_1 -v_1
We then notice that if the - is used, v_2 = v_1 which does not make sense which is why we use the smaller root (-)
As shown, this means that V_2 = 2GM /(v_1 * l_1) - V_1'''

#Part b)

sun_mass = 1.9891e30
G = 6.6738e-11

#Collect required information to solve for other paramters

L1 = eval(input('Enter perihelion distance from Sun (m): '))
V1 = eval(input('Enter velocity at perihelion (m/s): '))

# Solve for all quantities ℓ2, v2, T, and e using the 5 given equations in the textbook 

def V2(v1,l1):
     return -v1 + ((2 * G * sun_mass)/(v1*l1))

V_A = V2(V1,L1)
print(f'The velocity at the aphelion is: {V_A}', 'm/s')


def L2(l1,v1,v2):
    return (l1*v1/v2)

L_A = L2(L1,V1,V_A)
print(f'The distance at the aphelion is: {L_A}', 'm')


def Semi_Maj(l1,l2):
    return (1/2)*(l1 + l2)

a1 = Semi_Maj(L1, L_A)
    
def Semi_Min(l1, l2):
    return np.sqrt(l1 * l2)

b1 = Semi_Min(L1, L_A)

def Orbital_Period(a,b,c,d):
    return ((2 * np.pi * a * b)/(c * d))


print(f'The Orbital Period is: {Orbital_Period(a1 , b1, L1, V1)}', 's')

def Orbital_Eccen(a,b):
    return ((b - a)/(b + a))

print(f'The Orbital Eccentricity is: {Orbital_Eccen(L_A, L1)}')


print('')

#Part c)

#Calculate properties of Earth and Halley's Comet

#Perihelion velocity and distance for Earth denoted by E 

LE = 1.4710e11
VE = 3.0287e4

VAE = V2(VE,LE) #Calculate velocity at aphelion
print(f'The velocity at the aphelion for Earth is: {VAE}', 'm/s')

LAE = L2(LE, VE, VAE) #Calculate distance at aphelion
print(f'The distance at the aphelion for Earth is: {LAE}', 'm')

a2 = Semi_Maj(LE, LAE) #Calculate semi and major axes 
b2 = Semi_Min(LE, LAE)

print(f'The Orbital Period of Earth is: {Orbital_Period(a2,b2, LE, VE)}', 's') #Calculate Orbital period using function


print(f'The Orbital Eccentricity for Earth is: {(Orbital_Eccen(LE,LAE))}') #Calculate orbital eccentricity using function

print('')

#Repeat for Halley's Comet using the same functions and different distance and velocity at perihelion

LH = 8.7830e10
VH = 5.4529e4

VAH = V2(VH,LH)
print(f'The velocity at the aphelion for Halley Comet is: {VAH}', 'm/s')

LAH = L2(LH, VH, VAH)
print(f'The distance at the aphelion for Halley Comet is: {LAH}', 'm')

a3 = Semi_Maj(LH, LAH)
b3 = Semi_Min(LH, LAH)

print(f'The Orbital Period of Halley Comet is: {Orbital_Period(a3, b3, LH, VH)}', 's')

print(f'The Orbital Eccentricity for Halley Comet is: {Orbital_Eccen(LH,LAH)}')

#As shown below, both the Orbital Period for Earth and Halley's comet is correct 
```
